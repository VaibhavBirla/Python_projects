```python
import pandas as pd
import numpy as py
```

```python
df=pd.read_csv("songs_normalize.csv")
```

```python
df_copy=df #takinga copy of data
```

```python
df_copy[['genre1','genre2','genre3','genre4',]]=df_copy['genre'].str.split(",",expand=True) #splitting genres into 1 major genre
```

```python
drop_cols=["genre","genre2","genre3","genre4"]
```

```python
df_copy=df_copy.drop(columns=drop_cols)
```

```python
df_copy['genre']=df_copy['genre1']
```

```python
df_copy=df_copy.drop(columns="genre1")
```

```python
df_copy["Duration_min"]=df_copy["duration_ms"]/100000  #standardizing duration by converting it into minutes
```

```python
df_copy=df_copy.drop(columns='duration_ms')
```

```python
spotify_genre_populatity=df_copy.groupby("genre")["popularity"].mean().reset_index().sort_values(by="popularity",ascending=False)
```

```python
spotify_genre_populatity            #most popular genres
```

```python
spotify_genre_populatity_artists=df_copy.groupby("artist")["song"].count().reset_index().sort_values(by="song",ascending=False)
```

```python
spotify_genre_populatity_artists.head()      #artists with most hits
```
